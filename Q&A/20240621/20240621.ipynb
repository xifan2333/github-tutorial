```python

#! 2. ä½¿ç”¨å•ä¾‹æ¨¡å¼
import pandas as pd

class Params:
    def __new__(cls):
        if not hasattr(cls, "instance"):
            print("ä»excelä¸­è½½å…¥å‚æ•°")
            data = pd.read_excel("params.xlsx")
            cls.instance = super().__new__(cls)
            cls.instance.name = data["name"][0]
            cls.instance.age = int(data["age"][0])
            cls.instance.gender = data["gender"][0]
        return cls.instance

params = Params()


print(f"å§“åï¼š{params.name}")
print(f"å¹´é¾„ï¼š{params.age}")
print(f"æ€§åˆ«:{params.gender}")

#? ç¬¬äºŒæ¬¡è°ƒç”¨Params()æ—¶ï¼Œè¿”å›çš„æ˜¯ç¬¬ä¸€æ¬¡è°ƒç”¨æ—¶åˆ›å»ºçš„å¯¹è±¡
params2 = Params()
print(params is params2) # True
```

```python

#! 3. ä½¿ç”¨è£…é¥°å™¨
import pandas as pd
import functools

def provide_params(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        if not hasattr(wrapper, "params"):
            print("ä»excelä¸­è½½å…¥å‚æ•°")
            data = pd.read_excel("params.xlsx")
            wrapper.params = {
                "name": data["name"][0],
                "age": int(data["age"][0]),
                "gender":data["gender"][0]
                }
        return func(*args, **kwargs)
    return wrapper

@provide_params
def show_params():
    print(f"å§“åï¼š{show_params.params['name']}")
    print(f"å¹´é¾„ï¼š{show_params.params['age']}")
    print(f"æ€§åˆ«ï¼š{show_params.params['gender']}")

show_params()

#? ç¬¬äºŒæ¬¡è°ƒç”¨show_params()æ—¶ï¼Œä½¿ç”¨çš„æ˜¯wrapperçš„paramså±æ€§ï¼Œè€Œä¸æ˜¯é‡æ–°è½½å…¥å‚æ•°

show_params()



```
